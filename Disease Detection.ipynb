```python
import zipfile
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the path to the zip file
zip_file_path = r'C:/Users/Anjali/Downloads/medicine_dataset.csv.zip'



# Unzipping the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(r'C:/Users/Anjali/Downloads')

# Load the extracted CSV file into a DataFrame
csv_file_path = r'C:/Users/Anjali/Downloads/medicine_dataset.csv'
data = pd.read_csv(csv_file_path)


print(data.head())


print(data.describe())


print(data.isnull().sum())


plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Category', order=data['Category'].value_counts().index)
plt.title('Distribution of Drug Categories')
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()


plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Category', y='Strength')
plt.title('Strength of Drugs by Category')
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Strength')
plt.show()


plt.figure(figsize=(12, 6))
sns.countplot(data=data, x='Dosage Form', order=data['Dosage Form'].value_counts().index)
plt.title('Distribution of Dosage Forms')
plt.xticks(rotation=45)
plt.xlabel('Dosage Form')
plt.ylabel('Count')
plt.show()


```
